```python
source_file = 'jw300.' + source_language 
target_file = 'jw300.' + target_language
french_test = {}
source = []
target = []
english_sentences_in_global_test_set = {} 
with open(source_file) as src_f:
    for i, line in enumerate(src_f):
            source.append(line.strip())      


with open(target_file) as f:
    for j, line in enumerate(f):
            target.append(line.strip())
    
    
df = pd.DataFrame(zip(source, target), columns=['fr_sentence', 'mos_sentence'])
df.head(10)
```

### Save table

```python
path = '/content/drive/My Drive/Colab Notebooks/TraMoore/data_test/'
def save_file(array, file_name):
    with open(path + file_name, "w+", encoding="utf-8") as file:
        joined_array = "\n".join(array)
        file.write(joined_array)
```

```python
save_file(source, "fr-mos.fr") # mos
save_file(target, "fr-mos.mos") # fr
df.head(100).to_csv(path + 'fr-mos.csv') # fr-mos
```
